In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

# The std provided in the np array made no sense, was way too big, I reduced it by 100X
df = pd.DataFrame([np.random.normal(32000,2000,3650), 
                   np.random.normal(43000,1000,3650), 
                   np.random.normal(43500,1400,3650), 
                   np.random.normal(48000,700,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
# quick example of data in 5 point summary
df.T.describe()
```

```python
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.figure()
```

```python
# dataframe of aggregates and range of two standard deviations (95% conf interval) per year
# used to evaluated what color to turn the bar, based on y-axis interaction
dfT = df.T

dfAgg = pd.DataFrame({'mean' : dfT.mean().tolist(),
              'std' : dfT.std().tolist()},
             index=['1992','1993','1994','1995'])

dfAgg['-2 std'] = dfAgg['mean'] - dfAgg['std']*2
dfAgg['+2 std'] = dfAgg['mean'] + dfAgg['std']*2
dfAgg.reset_index()
dfAgg
```

```python
# plot mean as a bar graph and std as and error bar as two standard deviations from mean
plt.gca().cla()
plt.gca().set_title('Assignment 3 Figure')
# create means and stds as np.array() instead of lists, so math can be done more easily
means = np.array(dfAgg['mean'].tolist())
stds = np.array(dfAgg['std'].tolist())
plt.bar(range(1,5), means, yerr=stds*2, align='center', alpha=.75, capsize=10, picker=10, edgecolor='k')
plt.gca().set_xticks(range(1,5))
plt.gca().set_xticklabels(labels=['1992','1993','1994','1995'])
plt.ylim(bottom=0, top=60000)
```

```python
# pick a valid value from axes 
# this method is not employed in this assignment
def onpick(event):
    y = event.mouseevent.ydata
    plt.gca().axhline(y, color='g')
    plt.draw()

# tell mpl_connect we want to pass a 'pick_event' into onpick when the event is detected
#plt.gcf().canvas.mpl_connect('pick_event', onpick)
```

```python
# press anywhere in figure

def onpress(event):
    # clear the gray lines
    for line in plt.gca().lines :
        if line.get_c() == 'gray' :
            line.remove()
        else :
            continue
    
    # x, y coord and draw horizontal gray line
    x = event.xdata
    y = event.ydata
    
    # create a list of our conditions
    conditions = [
        (dfAgg['-2 std'] <= y) & (dfAgg['+2 std'] >= y),
        (dfAgg['+2 std'] < y),
        (dfAgg['-2 std'] > y)
        ]

    # create a list of the values we want to assign for each condition
    values = ['w','b','r']

    # create a new column and use np.select to assign values to it using our lists as arguments
    dfAgg['color'] = np.select(conditions, values)
    colors = dfAgg['color'].tolist()
    
    # set_color of bars in order 
    for child, color in zip(plt.gca().containers[1].get_children(), colors) :
        child.set_color(color)
        child.set_edgecolor('k')
    
    # put on gray line
    plt.gca().axhline(y, color='gray',alpha=.5)

plt.gcf().canvas.mpl_connect('button_press_event' ,onpress)
```
